```python
track_dict = defaultdict(int)

with open('./spotify_playlists_dataset/SPD_modified.tsv', mode = 'r') as f:

    lines = f.readlines()

for line in lines[1:]:
    data = line.split('\t,\t')

    track = (data[1],data[2])

    if track not in track_dict:
        track_dict[track] = len(track_dict)
```

SPDからユーザ-プレイリストのデータへ

```python
#user and playlist pair dict
tmp_list = []
user_playlist_pairs = defaultdict(list)
user = ''
playlist = ''

with open('./spotify_playlists_dataset/SPD_modified.tsv', mode = 'r') as f:

    lines = f.readlines()


with open('./spotify_playlists_dataset/user_playlist_pairs.tsv', mode='w') as f:
    for line in tqdm(lines[1:]):

        data = line[:-1].split('\t,\t')
        if user != data[0] or playlist != data[3] :
            f.write('\n')
            user = data[0]
            playlist = data[3]
            f.write(user+'\t,\t'+playlist)
        
        track = ( data[1], data[2] )
        f.write('\t,\t'+str(track_dict[track]))
```

ユーザid-プレイリストのデータからユーザid-トラックid-label(1)のデータへ

```python
with open('./spotify_playlists_dataset/user_playlist_pairs.tsv', mode = 'r') as f:

    lines = f.readlines()


with open('./spotify_playlists_dataset/session_data.tsv', mode = 'w') as f:
    f.write('userId\ttrackId\tlabel\n')

    for line in tqdm(lines[1:]):
        data = line.split('\t,\t')
        user = data[0]
        session = data[2:]

        for track in session:
            f.write( user+'\t'+track.removesuffix('\n')+'\t'+str(1)+'\n' )
```
