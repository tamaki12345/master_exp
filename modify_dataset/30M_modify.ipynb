```python
track_dict = defaultdict(tuple)

with open( '/home/tamak/master_exp/ThirtyMusic/entities/tracks.idomaar', mode = 'r' ) as f:
    lines = f.readlines()

    for line in lines:
        try:
           entitiy_type, track_id, _, info, extra_info = line.split('\t')
        except:
            print(line)
            for l in line.split():
                print(l)
            break

        
        name = json.loads(info)['name']
        artistname, trackname = name.split('/_/')

        artistname = unquote(artistname.replace('+',' '))
        trackname = unquote(trackname.replace('+',' '))

        if track_id not in track_dict:
            track_dict[ track_id ] = (artistname, trackname)
```

```python
SPD_set = set()

with open('./spotify_playlists_dataset/SPD_modified.tsv', mode = 'r') as f:

    lines = f.readlines()

for line in lines[1:]:
    data = line.split('\t,\t')

    if data[1] != '' and data[2] != '':

        track = (data[1],data[2])

        SPD_set.add( track )
```

```python
SPD_dict = defaultdict(int)

for track in SPD_set:
    SPD_dict[track] = len(SPD_dict)
```

# make test dataset

ユーザ-楽曲のデータのうち共有された楽曲に絞り込み

```python
modify_list = []

with open('30M_dataset/same_duration_data.tsv', mode = 'r') as f:
    lines  = f.readlines()
    print(len(lines))

    for line in lines[1:]:
        user_id, track_id = line.split()

        track = track_dict[track_id]

        if track in SPD_set:
            modify_list.append((user_id, SPD_dict[track]))
```

```python
len(modify_list)
```

```python
with open('30M_dataset/test_table_data.tsv', mode = 'w') as f:
    f.write('userId\ttrackId\n')
    for user_id, track_id, in modify_list:
        f.write(str(user_id)+'\t'+str(track_id)+'\n')
```

# same_duration_dataをsessionにしたので再生成

```python
modify_list = []

with open('30M_dataset/same_duration_data.tsv', mode = 'r') as f:
    lines  = f.readlines()
    print(len(lines))

    for line in lines[1:]:
        tmp = []
        data = line.split('\t')

        user_id = data[0]
        session = data[1:]

        tmp.append(user_id)

        for track_id in session:
            track = track_dict[track_id]

            if track in SPD_set:
                tmp.append(SPD_dict[track])

        if len(tmp) > 1:
            modify_list.append(tmp)
```

```python
with open('30M_dataset/test_session_data.tsv', mode = 'w') as f:
    f.write('userId\tsession\n')
    for data in modify_list:
        
        user_id = data[0]
        session = data[1:]

        f.write(str(user_id))
        
        for track_id in session:
            f.write('\t'+str(track_id))
        f.write('\n')
```

```python
mdfy_list = []

with open('30M_dataset/test_session_data.tsv', mode = 'r') as f:

    print(len(f.readlines()))
    
    for session_id, line in enumerate(f.readlines()[1:]):
        user_id = line[0]
        session = line[1:].split('\t')

        for track_id in session:
            mdfy_list.append( ( str(session_id), str(user_id), str(track_id.rstrip('\n')) ) )

with open('30M_dataset/experiment_test.txt', mode = 'w') as f:
    f.write('SessionId\tItemId\n')

    for data in mdfy_list:
        session_id, user_id, track_id = data
        f.write(session_id+'\t'+track_id+'\n')
```
